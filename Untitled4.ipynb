```python
# Import necessary libraries
import spacy  # For Natural Language Processing (NLP)
import networkx as nx  # For building the Knowledge Graph (KG)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score  # Evaluation metrics
import matplotlib.pyplot as plt  # For visualizing confusion matrix
import random  # For random operations like shuffling and sampling

# Load spaCy NLP models for English and German
nlp_en = spacy.load('en_core_web_sm')  # English model for POS tagging and tokenization
nlp_de = spacy.load('de_core_news_sm')  # German model for tokenization and POS tagging (optional)

# Initialize the Knowledge Graph
G = nx.Graph()  # Create an empty undirected graph (could be directed if needed)

# Function to add a translation pair to the Knowledge Graph
def add_translation(en_word, de_word, en_pos=None, de_pos=None):
    # Normalize words to lowercase for consistency
    en_word = en_word.lower()
    de_word = de_word.lower()

    # Add English word node with language and part-of-speech (POS)
    G.add_node(en_word, lang='en', pos=en_pos)

    # Add German word node with language and POS (optional)
    G.add_node(de_word, lang='de', pos=de_pos)

    # Create an edge between English and German words with a 'translation' relationship
    G.add_edge(en_word, de_word, relation='translation')

# Function to parse a line from the dataset and extract sentences
def process_line(line):
    # Dataset format: English sentence \t German sentence \t Metadata
    parts = line.strip().split('\t')

    # Ensure the line has at least two parts: English and German sentences
    if len(parts) < 2:
        return None, None

    en_sentence = parts[0].strip()  # Extract English sentence
    de_sentence = parts[1].strip()  # Extract German sentence

    return en_sentence, de_sentence

# Function to process sentence pairs and add them to the Knowledge Graph
def process_sentence_pair(en_sentence, de_sentence):
    en_doc = nlp_en(en_sentence)  # NLP processing for English sentence
    de_doc = nlp_de(de_sentence)  # NLP processing for German sentence

    # Basic alignment: pair words by index (not accurate for complex sentences!)
    for i, en_token in enumerate(en_doc):
        if en_token.is_punct:  # Skip punctuation
            continue

        if i >= len(de_doc):  # Ensure we don't go out of bounds
            break

        de_token = de_doc[i]  # Get the corresponding German token

        # Add translation to KG with POS tags
        add_translation(en_token.text, de_token.text, en_token.pos_, de_token.pos_)

# Function to translate an English sentence using the Knowledge Graph
def translate_sentence(en_sentence):
    en_doc = nlp_en(en_sentence)  # NLP processing for English sentence
    translated_sentence = []  # Store translated German words

    for token in en_doc:
        if token.is_punct:  # Skip punctuation
            continue

        # Translate each English word by looking it up in the KG
        word_translation = translate_word(token.text.lower())

        # Append translation (or placeholder if not found)
        translated_sentence.append(word_translation)

    return translated_sentence

# Function to translate a single English word using the Knowledge Graph
def translate_word(en_word):
    if en_word in G:
        # Retrieve all connected nodes (neighbors)
        neighbors = G[en_word]

        for neighbor in neighbors:
            if G.nodes[neighbor]['lang'] == 'de':  # Look for the German node
                return neighbor

    # Return the original word in brackets if no translation is found
    return f"[{en_word}]"

# Function to evaluate translations on test data and display confusion matrix and accuracy
def evaluate_translations(test_data):
    y_true = []  # Ground truth German words
    y_pred = []  # Predicted German words

    # Iterate over each sentence pair in the test set
    for en_sentence, de_sentence in test_data:
        # Clean and split German ground truth sentence into words
        ground_truth = de_sentence.replace('.', '').replace('!', '').replace('?', '').split()

        # Get the translated sentence from our KG model
        prediction = translate_sentence(en_sentence)

        # Display translation result for the current sentence
        #print(f"\nEN: {en_sentence}")
        #print(f"Ground Truth: {ground_truth}")
        #print(f"Predicted   : {prediction}")

        # Append words to true/pred lists (one-to-one alignment)
        for gt_word, pred_word in zip(ground_truth, prediction):
            y_true.append(gt_word.lower())
            y_pred.append(pred_word.lower())

    # # Create confusion matrix labels (unique set of all true and predicted words)
    # labels = sorted(list(set(y_true + y_pred)))

    # # Compute confusion matrix
    # cm = confusion_matrix(y_true, y_pred, labels=labels)

    # # Calculate accuracy score (percentage of correctly predicted words)
    accuracy = accuracy_score(y_true, y_pred) * 100  # Convert to percentage

    # # Display confusion matrix
    # disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    # disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)

    # plt.title("Confusion Matrix: Word-Level Translation")
    # plt.show()

    # Print the test accuracy result
    print(f"\n✅ Test Accuracy: {accuracy:.2f}%")

# Function to simulate random predictions on unseen sentences
def random_predictions(num_predictions=5):
    # Sample random English sentences (some may be unseen!)
    sample_sentences = [
        "Go!",
        "Run!",
        "Hi!",
        "He runs.",
        "I eat.",
        "You run.",
        "We play.",
        "She eats.",
        "Drive fast.",
        "Stop now!"
    ]

    print("\n🔮 Random Predictions on New Sentences:")

    # Randomly pick sentences and predict their translations
    for _ in range(num_predictions):
        sentence = random.choice(sample_sentences)
        prediction = translate_sentence(sentence)

        print(f"EN: {sentence}")
        print(f"Predicted DE: {prediction}\n")

# Main pipeline function that runs everything
def test_pipeline():
    # ✅ Step 1: Load dataset from your clean text file
    dataset_file = '/content/drive/MyDrive/Classroom/deu.txt'  # Replace with your actual filename
    dataset_lines = []

    print(f"\n📂 Reading dataset from: {dataset_file}")

    # Read all lines first (even invalid)
    with open(dataset_file, 'r', encoding='utf-8') as f:
        all_lines = f.readlines()
    all_lines = all_lines[:100000]
    print(f"📋 Total lines in file (raw): {len(all_lines)}")

    # Now filter for lines with valid sentence pairs
    for line in all_lines:
        en_sentence, de_sentence = process_line(line)
        if en_sentence and de_sentence:
            dataset_lines.append(line)

    print(f"✅ Valid dataset lines (sentence pairs): {len(dataset_lines)}")

    # Safety check: if dataset_lines is still small
    if len(dataset_lines) < 10:
        print("⚠️ WARNING: Very few valid lines. Check your file structure or parsing logic.")
        return  # Stop the function if no good data

    # ✅ Step 2: Shuffle and split dataset into 60% training and 40% testing
    random.shuffle(dataset_lines)
    split_idx = int(0.6 * len(dataset_lines))

    train_lines = dataset_lines[:split_idx]
    test_lines = dataset_lines[split_idx:]

    print(f"\n🛠️  Training on {len(train_lines)} sentence pairs.")
    print(f"🔬 Testing on {len(test_lines)} sentence pairs.")

    # ✅ Step 3: Build Knowledge Graph from the training data
    for line in train_lines:
        en_sentence, de_sentence = process_line(line)
        if en_sentence and de_sentence:
            process_sentence_pair(en_sentence, de_sentence)

    print("\n✅ Knowledge Graph Training Completed.")

    # ✅ Step 4: Prepare test data (English and German sentence pairs)
    test_data = []
    for line in test_lines:
        en_sentence, de_sentence = process_line(line)
        if en_sentence and de_sentence:
            test_data.append((en_sentence, de_sentence))

    # ✅ Step 5: Evaluate translations on the test set
    evaluate_translations(test_data)

    # ✅ Step 6: Show sample nodes and edges from the Knowledge Graph
    print("\n🧠 Knowledge Graph Nodes (sample):")
    for node, data in list(G.nodes(data=True))[:10]:  # Show first 10 nodes
        print(f"{node}: {data}")

    print("\n🧠 Knowledge Graph Edges (translations sample):")
    for src, tgt, data in list(G.edges(data=True))[:10]:  # Show first 10 edges
        print(f"{src} -> {tgt}: {data}")

    # ✅ Step 7: Perform random predictions on new sentences
    random_predictions(num_predictions=5)

# Run the pipeline
test_pipeline()
```

```python
# ✅ Install requirements in Colab first (uncomment these lines when running in Colab)
# !pip install spacy tensorflow networkx
# !python -m spacy download en_core_web_sm
# !python -m spacy download de_core_news_sm

# ✅ Imports
import numpy as np
import tensorflow as tf
import networkx as nx
import spacy
import re
import unicodedata
import random
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

# ✅ Load spaCy models for NLP
nlp_en = spacy.load('en_core_web_sm', disable=["ner", "parser"])
nlp_de = spacy.load('de_core_news_sm', disable=["ner", "parser"])

# ✅ Initialize Knowledge Graph
G = nx.Graph()

# ✅ Function: Add nodes and edges to the KG
def add_to_knowledge_graph(en_word, de_word, en_pos=None, de_pos=None):
    en_word = en_word.lower()
    de_word = de_word.lower()
    G.add_node(en_word, lang='en', pos=en_pos)
    G.add_node(de_word, lang='de', pos=de_pos)
    G.add_edge(en_word, de_word, relation='translation')

# ✅ Function: Process sentences and add them to KG
def process_to_kg(en_sentence, de_sentence):
    en_doc = nlp_en(en_sentence)
    de_doc = nlp_de(de_sentence)
    for i, en_token in enumerate(en_doc):
        if en_token.is_punct:
            continue
        if i >= len(de_doc):
            break
        de_token = de_doc[i]
        add_to_knowledge_graph(en_token.text, de_token.text, en_token.pos_, de_token.pos_)

# ✅ Preprocessing functions
def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                   if unicodedata.category(c) != 'Mn')

def preprocess_sentence(s):
    s = unicode_to_ascii(s.lower().strip())
    s = re.sub(r"([?.!,¿])", r" \1 ", s)
    s = re.sub(r'[" "]+', " ", s)
    s = re.sub(r"[^a-zA-Z?.!,¿]+", " ", s)
    s = s.strip()
    return s

# ✅ Load dataset and build KG + dataset for NMT
data_path = '/content/drive/MyDrive/Classroom/deu.txt'  # ✅ Your dataset path
num_samples = 100000  # ✅ Adjust the number for testing

input_texts = []
target_texts = []

with open(data_path, 'r', encoding='utf-8') as f:
    lines = f.read().split('\n')

for line in lines[:num_samples]:
    parts = line.split('\t')
    if len(parts) < 2:
        continue
    en_sentence = preprocess_sentence(parts[0])
    de_sentence = preprocess_sentence(parts[1])

    input_texts.append(en_sentence)
    target_texts.append(f"<start> {de_sentence} <end>")

    # ✅ Build KG from the original (non-preprocessed) sentences
    process_to_kg(parts[0], parts[1])

print(f"✅ Loaded {len(input_texts)} sentence pairs.")
print(f"✅ Knowledge Graph has {len(G.nodes)} nodes and {len(G.edges)} edges.")

# ✅ Tokenize and prepare data for NMT
input_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
input_tokenizer.fit_on_texts(input_texts)
input_sequences = input_tokenizer.texts_to_sequences(input_texts)
input_max_len = max(len(seq) for seq in input_sequences)
input_vocab_size = len(input_tokenizer.word_index) + 1

target_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
target_tokenizer.fit_on_texts(target_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)
target_max_len = max(len(seq) for seq in target_sequences)
target_vocab_size = len(target_tokenizer.word_index) + 1

encoder_input_data = tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=input_max_len, padding='post')
decoder_input_data = tf.keras.preprocessing.sequence.pad_sequences(target_sequences, maxlen=target_max_len, padding='post')
decoder_target_data = np.zeros_like(decoder_input_data)
decoder_target_data[:, :-1] = decoder_input_data[:, 1:]

# ✅ NMT Model (Encoder-Decoder with LSTM)
latent_dim = 256
batch_size = 64
epochs = 100

# Encoder
encoder_inputs = tf.keras.Input(shape=(None,))
enc_emb = tf.keras.layers.Embedding(input_vocab_size, latent_dim)(encoder_inputs)
encoder_outputs, state_h, state_c = tf.keras.layers.LSTM(latent_dim, return_state=True)(enc_emb)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = tf.keras.Input(shape=(None,))
dec_emb_layer = tf.keras.layers.Embedding(target_vocab_size, latent_dim)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = tf.keras.layers.LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = tf.keras.layers.Dense(target_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Full training model
model = tf.keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
decoder_target_data = np.expand_dims(decoder_target_data, -1)

print("✅ Training NMT model...")
model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
          batch_size=batch_size, epochs=epochs, validation_split=0.2)

# ✅ Inference Models
encoder_model = tf.keras.Model(encoder_inputs, encoder_states)

decoder_state_input_h = tf.keras.Input(shape=(latent_dim,))
decoder_state_input_c = tf.keras.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

dec_emb_inf = dec_emb_layer(decoder_inputs)
decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm(
    dec_emb_inf, initial_state=decoder_states_inputs)
decoder_outputs_inf = decoder_dense(decoder_outputs_inf)
decoder_states_inf = [state_h_inf, state_c_inf]

decoder_model = tf.keras.Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs_inf] + decoder_states_inf)

# ✅ Inference Function
def decode_sequence(input_seq):
    states_value = encoder_model.predict(input_seq)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = target_tokenizer.word_index['<start>']

    decoded_sentence = ''
    stop_condition = False

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])

        sampled_word = None
        for word, index in target_tokenizer.word_index.items():
            if index == sampled_token_index:
                sampled_word = word
                break

        if sampled_word == '<end>' or len(decoded_sentence.split()) > target_max_len:
            stop_condition = True
        else:
            decoded_sentence += ' ' + sampled_word

        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        states_value = [h, c]

    return decoded_sentence.strip()

# ✅ Run predictions and compare with KG outputs
print("\n✅ Random Predictions + Knowledge Graph Cross-Check:\n")
for seq_index in range(5):
    input_seq = encoder_input_data[seq_index: seq_index + 1]
    decoded_sentence = decode_sequence(input_seq)
    original_en = input_texts[seq_index]

    # KG-based translation (simple word-by-word)
    en_tokens = nlp_en(original_en)
    kg_translation = []
    for token in en_tokens:
        word = token.text.lower()
        if word in G:
            neighbors = G[word]
            de_word = None
            for n in neighbors:
                if G.nodes[n]['lang'] == 'de':
                    de_word = n
                    break
            kg_translation.append(de_word if de_word else f"[{word}]")
        else:
            kg_translation.append(f"[{word}]")

    print(f"EN Input      : {original_en}")
    print(f"NMT Translation: {decoded_sentence}")
    print(f"KG Translation : {' '.join(kg_translation)}")
    print("-" * 60)




smoothie = SmoothingFunction().method4

def evaluate_bleu(test_samples=100):
    total_bleu = 0
    count = 0

    for seq_index in range(test_samples):
        input_seq = encoder_input_data[seq_index: seq_index + 1]
        decoded_sentence = decode_sequence(input_seq)

        # Reference (Ground Truth)
        reference = target_texts[seq_index].replace('<start>', '').replace('<end>', '').strip()
        reference_tokens = nltk.word_tokenize(reference)

        # Prediction
        candidate_tokens = nltk.word_tokenize(decoded_sentence)

        # BLEU score for this sample
        bleu_score = sentence_bleu([reference_tokens], candidate_tokens, smoothing_function=smoothie)
        total_bleu += bleu_score
        count += 1

        if seq_index < 5:  # Print first few examples
            print('-' * 60)
            print(f"EN Input       : {input_texts[seq_index]}")
            print(f"Ground Truth   : {reference}")
            print(f"NMT Translation: {decoded_sentence}")
            print(f"BLEU Score     : {bleu_score:.4f}")

    avg_bleu = total_bleu / count
    print("\n✅ Average BLEU Score on Test Samples:", round(avg_bleu * 100, 2), "%")

# Run BLEU evaluation
evaluate_bleu(test_samples=100)
```

```python
# ✅ Install requirements in Colab first (uncomment these lines when running in Colab)
# !pip install spacy tensorflow networkx nltk matplotlib
# !python -m spacy download en_core_web_sm
# !python -m spacy download de_core_news_sm

# ✅ Imports
import numpy as np
import tensorflow as tf
import networkx as nx
import spacy
import re
import unicodedata
import random
from collections import defaultdict
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate import meteor_score
from nltk import word_tokenize
import nltk
from google.colab import drive

drive.mount('/content/drive')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')  # For METEOR score

# ✅ Load spaCy models for NLP
nlp_en = spacy.load('en_core_web_sm', disable=["ner"])
nlp_de = spacy.load('de_core_news_sm', disable=["ner"])

# ✅ Initialize Knowledge Graph
G = nx.DiGraph()  # Directed graph for better relation tracking

# ✅ Function: Add nodes and edges to the Knowledge Graph
def add_to_knowledge_graph(en_word, de_word, en_pos=None, de_pos=None, weight=1.0):
    en_word = en_word.lower()
    de_word = de_word.lower()
    if en_word not in G:
        G.add_node(en_word, lang='en', pos=en_pos)
    if de_word not in G:
        G.add_node(de_word, lang='de', pos=de_pos)
    if G.has_edge(en_word, de_word):
        G[en_word][de_word]['weight'] += weight
    else:
        G.add_edge(en_word, de_word, relation='translation', weight=weight)

# ✅ Function: Process sentences and build KG
def process_to_kg(en_sentence, de_sentence):
    en_doc = nlp_en(en_sentence)
    de_doc = nlp_de(de_sentence)
    en_tokens = [token for token in en_doc if not token.is_punct]
    de_tokens = [token for token in de_doc if not token.is_punct]
    min_len = min(len(en_tokens), len(de_tokens))
    for i in range(min_len):
        add_to_knowledge_graph(en_tokens[i].text, de_tokens[i].text, en_tokens[i].pos_, de_tokens[i].pos_)

# ✅ Preprocessing functions
def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')

def preprocess_sentence(s):
    s = unicode_to_ascii(s.lower().strip())
    s = re.sub(r"([?.!,¿])", r" \1 ", s)
    s = re.sub(r'[" "]+', " ", s)
    s = re.sub(r"[^a-zA-Z?.!,¿]+", " ", s)
    return s.strip()

# ✅ Load dataset and prepare data
data_path = '/content/drive/MyDrive/Classroom/deu.txt'   # Adjust path as needed
num_samples = 10000  # Reduced for demo; increase for thesis

input_texts = []
target_texts = []
kg_pairs = []

with open(data_path, 'r', encoding='utf-8') as f:
    lines = f.read().strip().split('\n')

for line in lines[:num_samples]:
    parts = line.split('\t')
    if len(parts) < 2:
        continue
    en_sentence, de_sentence = parts[0], parts[1]
    input_texts.append(preprocess_sentence(en_sentence))
    target_texts.append(f"<start> {preprocess_sentence(de_sentence)} <end>")
    kg_pairs.append((en_sentence, de_sentence))

# Build Knowledge Graph
for en_sent, de_sent in kg_pairs:
    process_to_kg(en_sent, de_sent)

print(f"✅ Loaded {len(input_texts)} sentence pairs.")
print(f"✅ Knowledge Graph: {len(G.nodes)} nodes, {len(G.edges)} edges.")

# ✅ Tokenization for NMT
input_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
input_tokenizer.fit_on_texts(input_texts)
input_sequences = input_tokenizer.texts_to_sequences(input_texts)
input_max_len = max(len(seq) for seq in input_sequences)
input_vocab_size = len(input_tokenizer.word_index) + 1

target_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
target_tokenizer.fit_on_texts(target_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)
target_max_len = max(len(seq) for seq in target_sequences)
target_vocab_size = len(target_tokenizer.word_index) + 1

# Padding sequences
# ✅ Tokenization and Padding (Updated)
max_seq_len = max(input_max_len, target_max_len)  # Use the larger of the two lengths

encoder_input_data = tf.keras.preprocessing.sequence.pad_sequences(
    input_sequences, maxlen=max_seq_len, padding='post'
)
decoder_input_data = tf.keras.preprocessing.sequence.pad_sequences(
    target_sequences, maxlen=max_seq_len, padding='post'
)
decoder_target_data = np.zeros_like(decoder_input_data)
decoder_target_data[:, :-1] = decoder_input_data[:, 1:]
decoder_target_data = np.expand_dims(decoder_target_data, -1)

print(f"✅ Unified max sequence length: {max_seq_len}")
print(f"Encoder input shape: {encoder_input_data.shape}")
print(f"Decoder input shape: {decoder_input_data.shape}")
print(f"Decoder target shape: {decoder_target_data.shape}")

latent_dim = 512
batch_size = 64
epochs = 50

# Encoder
encoder_inputs = tf.keras.Input(shape=(None,))
enc_emb = tf.keras.layers.Embedding(input_vocab_size, latent_dim, mask_zero=True)(encoder_inputs)
encoder_lstm = tf.keras.layers.LSTM(latent_dim, return_state=True, return_sequences=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]

# Decoder with Attention
decoder_inputs = tf.keras.Input(shape=(None,))
dec_emb_layer = tf.keras.layers.Embedding(target_vocab_size, latent_dim, mask_zero=True)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = tf.keras.layers.LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)

# Reshape encoder outputs to be compatible with attention layer
encoder_outputs = tf.keras.layers.Reshape((-1, latent_dim))(encoder_outputs) # Reshape to (batch_size, seq_len, latent_dim)


# Attention Layer - Ensure compatibility
attention = tf.keras.layers.Attention()
decoder_attended = attention([decoder_outputs, encoder_outputs])  # [query, value]

# Combine attended output with decoder output (optional, depending on your design)
decoder_combined = tf.keras.layers.Concatenate(axis=-1)([decoder_outputs, decoder_attended])
decoder_dense = tf.keras.layers.Dense(target_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_combined)

# Training Model
nmt_model = tf.keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)
nmt_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
nmt_model.summary()  # Print model architecture for debugging

# ✅ Save model at every epoch
checkpoint_filepath = '/content/drive/MyDrive/Classroom/nmt_model_epoch_{epoch:02d}.keras'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=False,
    monitor='val_loss',
    mode='min',
    save_best_only=False,
    verbose=1
)

# ✅ Train Model with Checkpoint
history = nmt_model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=0.2,
    callbacks=[model_checkpoint_callback]
)


# ✅ Inference Setup (Updated)
encoder_model = tf.keras.Model(encoder_inputs, [encoder_outputs] + encoder_states)

decoder_state_input_h = tf.keras.Input(shape=(latent_dim,))
decoder_state_input_c = tf.keras.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
dec_emb_inf = dec_emb_layer(decoder_inputs)
decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm(dec_emb_inf, initial_state=decoder_states_inputs)
decoder_attended_inf = attention([decoder_outputs_inf, encoder_outputs])
decoder_combined_inf = tf.keras.layers.Concatenate(axis=-1)([decoder_outputs_inf, decoder_attended_inf])
decoder_outputs_inf = decoder_dense(decoder_combined_inf)

decoder_model = tf.keras.Model(
    [decoder_inputs] + decoder_states_inputs + [encoder_outputs],
    [decoder_outputs_inf] + [state_h_inf, state_c_inf]
)

# ✅ NMT Inference Function (Unchanged, but ensure it works with new shapes)
def decode_sequence_nmt(input_seq):
    encoder_outputs, h, c = encoder_model.predict(input_seq, verbose=0)
    states_value = [h, c]
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = target_tokenizer.word_index['<start>']
    decoded_sentence = ''
    while len(decoded_sentence.split()) < max_seq_len:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value + [encoder_outputs], verbose=0)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = target_tokenizer.index_word.get(sampled_token_index, '')
        if sampled_word == '<end>':
            break
        decoded_sentence += ' ' + sampled_word
        target_seq[0, 0] = sampled_token_index
        states_value = [h, c]
    return decoded_sentence.strip()

# ✅ Knowledge Graph Translation Function
def translate_with_kg(en_sentence):
    en_doc = nlp_en(en_sentence)
    translation = []
    for token in en_doc:
        if token.is_punct:
            translation.append(token.text)
            continue
        word = token.text.lower()
        if word in G:
            neighbors = sorted(G[word].items(), key=lambda x: x[1]['weight'], reverse=True)
            for neighbor, data in neighbors:
                if data['relation'] == 'translation':
                    translation.append(neighbor)
                    break
            else:
                translation.append(f"[{word}]")
        else:
            translation.append(f"[{word}]")
    return ' '.join(translation)

# ✅ Hybrid Approach (KG + NMT)
def hybrid_translation(en_sentence):
    nmt_result = decode_sequence_nmt(encoder_input_data[input_texts.index(preprocess_sentence(en_sentence))][np.newaxis, :])
    kg_result = translate_with_kg(en_sentence)
    # Simple heuristic: Use KG for known words, NMT for structure
    kg_words = kg_result.split()
    nmt_words = nmt_result.split()
    hybrid = []
    for i, kg_word in enumerate(kg_words):
        if kg_word.startswith('[') and i < len(nmt_words):
            hybrid.append(nmt_words[i])
        else:
            hybrid.append(kg_word)
    return ' '.join(hybrid)

# ✅ Evaluation Metrics
smoothie = SmoothingFunction().method4

def wer(reference, hypothesis):
    ref_words = reference.split()
    hyp_words = hypothesis.split()
    d = np.zeros((len(ref_words) + 1, len(hyp_words) + 1))
    for i in range(len(ref_words) + 1):
        d[i, 0] = i
    for j in range(len(hyp_words) + 1):
        d[0, j] = j
    for i in range(1, len(ref_words) + 1):
        for j in range(1, len(hyp_words) + 1):
            if ref_words[i-1] == hyp_words[j-1]:
                d[i, j] = d[i-1, j-1]
            else:
                d[i, j] = min(d[i-1, j-1], d[i-1, j], d[i, j-1]) + 1
    return d[len(ref_words), len(hyp_words)] / len(ref_words)

def evaluate_models(test_samples=100):
    bleu_scores = {'NMT': [], 'KG': [], 'Hybrid': []}
    wer_scores = {'NMT': [], 'KG': [], 'Hybrid': []}
    meteor_scores = {'NMT': [], 'KG': [], 'Hybrid': []}

    for i in random.sample(range(len(input_texts)), test_samples):
        input_text = input_texts[i]
        reference = target_texts[i].replace('<start>', '').replace('<end>', '').strip()
        ref_tokens = word_tokenize(reference)

        #Ascendingly, the NMT prediction requires a batch input, so we expand dims
        input_seq = encoder_input_data[i][np.newaxis, :]

        # Predictions
        nmt_pred = decode_sequence_nmt(input_seq)
        kg_pred = translate_with_kg(input_text)
        hybrid_pred = hybrid_translation(input_text)

        # BLEU Scores
        for method, pred in [('NMT', nmt_pred), ('KG', kg_pred), ('Hybrid', hybrid_pred)]:
          # Tokenize the prediction before passing to meteor_score
          pred_tokens = word_tokenize(pred)  # Tokenize 'pred'
          bleu = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=smoothie)
          bleu_scores[method].append(bleu)
          wer_score = wer(reference, pred)
          wer_scores[method].append(wer_score)

          # Use pred_tokens for meteor_score and ref_tokens as reference
          meteor = meteor_score.meteor_score([ref_tokens], pred_tokens)  # Pass tokenized prediction
          meteor_scores[method].append(meteor)

        # Display first few examples
        if i < 5:
            print(f"\nEN Input: {input_text}")
            print(f"Reference: {reference}")
            print(f"NMT: {nmt_pred}")
            print(f"KG: {kg_pred}")
            print(f"Hybrid: {hybrid_pred}")

    # Summary Statistics
    for method in ['NMT', 'KG', 'Hybrid']:
        avg_bleu = np.mean(bleu_scores[method]) * 100
        avg_wer = np.mean(wer_scores[method]) * 100
        avg_meteor = np.mean(meteor_scores[method]) * 100
        print(f"\n{method} Performance:")
        print(f"Avg BLEU: {avg_bleu:.2f}%")
        print(f"Avg WER: {avg_wer:.2f}%")
        print(f"Avg METEOR: {avg_meteor:.2f}%")

    # Visualization
    plt.figure(figsize=(12, 6))
    for method, color in [('NMT', 'b'), ('KG', 'g'), ('Hybrid', 'r')]:
        plt.plot(bleu_scores[method], label=f'{method} BLEU', color=color, alpha=0.5)
    plt.legend()
    plt.title('BLEU Score Distribution Across Methods')
    plt.xlabel('Sample Index')
    plt.ylabel('BLEU Score')
    plt.show()

# ✅ Run Evaluation
print("✅ Evaluating Models...")
evaluate_models(test_samples=100)
```
